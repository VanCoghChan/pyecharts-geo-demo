```python
import pyecharts
import pandas as pd
import numpy as np
import codecs
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv("./data.csv", encoding='gbk')
data.head()
```

```python
countries = data['国家'].tolist()
GDP_2016 = data['2016GDP(亿美元)'].tolist()
GDP_2017 = data['2017GPD(亿美元)'].tolist()
GDP_in = data['GDP增速'].tolist()
GDP_ave = data['人均GPD(美元)'].tolist()
pp = data['人口总数'].tolist()
GDP_in = [eval(x[:-1]) for x in GDP_in]
```

```python
from pyecharts import *
page = Page()

map0 = Map("2016年GDP统计", width=1400, height=500)
map0.add("2016年GDP统计", countries, GDP_2016, maptype="world",  is_visualmap=True,
         visual_text_color='#105', visual_range=[0, max(GDP_2016)])
bar0 = Bar("2016年GDP统计柱状图", width=1400, height=500, title_pos='center')
bar0.add("", countries, GDP_2016, is_more_utils=True, mark_point = ['max', 'min'])

map1 = Map("2017年GDP统计", width=1400, height=500)
map1.add("2017年GDP统计", countries, GDP_2017, maptype="world",  is_visualmap=True,
         visual_text_color='#105', visual_range=[0, max(GDP_2017)])
bar1 = Bar("2017年GDP统计柱状图", width=1400, height=500, title_pos='center')
bar1.add("", countries, GDP_2017, is_more_utils=True, mark_point = ['max', 'min'], label_color=['#FF3366'])

map2 = Map("GDP增长率统计", width=1400, height=500)
map2.add("GDP增长率统计", countries, GDP_in, maptype="world",  is_visualmap=True,
         visual_text_color='#105', visual_range=[0, max(GDP_in)])
line2 = Line("GDP增长率统计折线图", width=1400, height=500, title_pos='center')
line2.add("", countries, GDP_in, is_more_utils=True, mark_point = ['max', 'min'], label_color=['#4176AA'])

map3 = Map("人均GDP统计", width=1400, height=500)
map3.add("人均GDP统计", countries, GDP_ave, maptype="world",  is_visualmap=True,
         visual_text_color='#105', visual_range=[0, max(GDP_ave)])
line3 = Line("GDP增长率折线统计", width=1400, height=500, title_pos='center')
line3.add("", countries, GDP_ave, is_more_utils=True, mark_point = ['max', 'min'], label_color=['#CD5C5C'])

page.add(map0)
page.add(bar0)
page.add(map1)
page.add(bar1)
page.add(map2)
page.add(line2)
page.add(map3)
page.add(line3)

page.render('./sample.html')
# # map2 = Map("世界地图示例", width=2000, height=1000)
# # map2.add("世界地图", countries, GDP_in, maptype="world",  is_visualmap=True,
# #          visual_text_color='#105', visual_range=[0, max(GDP_in)])
# # bar2 = Bar("2017年GPD统计", title_pos="65%")
# # bar2.add("", countries, GDP_in)

# timeline = Timeline(is_auto_play=False, timeline_bottom=0, width=1300, height=700)
# timeline.add(page0, '2016年GDP统计')
# timeline.add(page1, '2017年GDP统计')
# timeline.add(map2, 'GDP增速')
```
